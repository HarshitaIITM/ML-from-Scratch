## working with numpy arrays

```python
t = np.arange(0,5,0.2)
#syntax 
# plt.plot(
#     x1, y1, format1,
#     x2, y2, format2,
#     x3, y3, format3
# )
plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^')
plt.show()
```

```python
#first create a grid of plots
#ax will be an array of two axes objects
#syntax
# fig, ax = plt.subplots()
# ax.plot(x, y)
#fig - figure object(whole canvas) and ax- axes objects (plotting areas)
fig,ax = plt.subplots(2)
#call plot() method on the appropriate object
ax[0].plot(x1,np.sin(x1),'b-')
ax[1].plot(x1,np.cos(x1),'b-')
```

```python
# Adding a legend -it used to describe what each line or curve menas in plot
# This method follows the MATLAB API. It is prone to errors and unflexible if curves are added to or removed from the plot. It resulted in a wrongly labelled curve.
x15 = np.arange(1,5)
fig,ax = plt.subplots()
ax.plot(x15,x15*1.5)
ax.plot(x15,x15*3.0)
ax.plot(x15,x15/3.0)
ax.legend(['Normal','Fast','slow'])
```

```python
#better method
x15 = np.arange(1, 5)

fig, ax = plt.subplots()

ax.plot(x15, x15*1.5, label='Normal')
ax.plot(x15, x15*3.0, label='Fast')
ax.plot(x15, x15/3.0, label='Slow')

ax.legend();
```
